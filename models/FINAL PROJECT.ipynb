```python
folderpath = "bitcoin_history_rates_apr21-jul21.csv"

dfbit = pd.read_csv(folderpath)
dfbit
```

```python

splitDate = dfbit[["time_period_start"]]
splitDate = np.array(splitDate)
date = []
time = []
for i in splitDate:
    datei = i[0]
    datei = datei.split("T")
    date.append(datei[0])
    time.append(datei[1])
    

dfbit["Date"] = date
data =dfbit["Date"] 
month = []
day = []
for i in data:
    datei = i.split("-")
    month.append(datei[1])
    day.append(datei[2])

dfbit["month"] = month
dfbit["day"] = day
dfbit = dfbit.drop(columns=["time_period_start","time_period_end","time_open","time_close"])
dfbit
```

```python
import matplotlib.pyplot as plt
highs = np.array(dfbit["rate_high"])
dates = np.array(dfbit["Date"])
import matplotlib.pyplot as plt
fig, axs = plt.subplots(figsize = (20,10))

show_every = 6

sparse_xticks = [None] * dates.shape[0]
sparse_xticks[::show_every] = dates[::show_every]

axs.set_title("BITCOIN HIGH VALUES FROM (April -June)", fontsize = 17.5)
axs.tick_params(axis = 'x', labelsize = 14)
axs.set_xlabel('dates', size = 15)
axs.tick_params(axis = 'y', labelsize =14)
axs.set_ylabel('price', size = 15)
axs.set_xticklabels(sparse_xticks, fontsize=14 ,rotation=45)
plt.plot(dates , highs)


plt.show()

SNLhost= "May 8"
Tesla_wont_take_bitcoin = "May-12-2021 5:08pm "


```
